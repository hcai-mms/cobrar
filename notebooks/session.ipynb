```python
import pandas as pd
import numpy as np
import os
import shutil
from util import create_sessions, delete_duplicates, k_core_filtering
```

```python
df_interactions = pd.read_csv('../data/userid_trackid_timestamp_interval_filtered.tsv', sep='\t') # intervall filtered
```

```python
df_emma = pd.read_csv('../data/id_emma.tsv', sep='\t')
df_emma_interactions = df_interactions[df_interactions['track_id'].isin(df_emma.id)].copy()
df_emma_interactions['track_id'].nunique()
```

```python
df_emma_interactions_filtered = create_sessions(df_emma_interactions)

df_emma_interactions_filtered = delete_duplicates(df_emma_interactions_filtered, user_col='session_id')

df_emma_interactions_filtered = k_core_filtering(df_emma_interactions_filtered, k_user=2, k_item=5, user_col='session_id')

print(df_emma_interactions_filtered.shape)
df_emma_interactions_filtered.sort_values(by=['session_id', 'timestamp'], inplace=True)
df_emma_interactions_filtered.head(10)
```

```python
df_emma_interactions_filtered.drop(columns=['user_id']).rename(columns={'session_id': 'user_id'}).to_csv('../data/userid_trackid_timestamp_interval_filtered_emma.tsv', sep='\t', index=False)
```

```python
df_interactions['year'] = pd.to_datetime(df_interactions['timestamp']).dt.to_period('Y')
df_interactions['month'] = pd.to_datetime(df_interactions['timestamp']).dt.to_period('M')
df_interactions['week'] = pd.to_datetime(df_interactions['timestamp']).dt.to_period('W')
#df_interactions = df_interactions[df_interactions['year'] == '2014']
df_interactions = df_interactions[df_interactions['week'] == '2014-03-03/2014-03-09']
print(df_interactions.shape)
df_interactions.head()
```

```python
df_interactions_filtered = create_sessions(df_interactions)

df_interactions_filtered = delete_duplicates(df_interactions_filtered, user_col='session_id')

df_interactions_filtered = k_core_filtering(df_interactions_filtered, k_user=2, k_item=5, user_col='session_id')

print(df_interactions_filtered.shape)
df_interactions_filtered.sort_values(by=['session_id', 'timestamp'], inplace=True)
df_interactions_filtered.head(10)
```

```python
df_interactions_filtered.drop(columns=['user_id']).rename(columns={'session_id': 'user_id'}).to_csv('../data/userid_trackid_timestamp_session_onion.tsv', sep='\t', index=False)
```
